## Distribution Plots
* jointplot
* pairplot
* distplot

```python
import seaborn as sns
```

```python
df = sns.load_dataset("tips") # dips is an inbuilt dataset
```

```python
df.head()
```

```python
# tip is a dependant feature and rest all are independant features
```

## Correlation with Heatmap

```python
df.corr() # will find the inter-relation between diff features (int and float features only) & range is -1 to +1
```

```python
# positive values above indicate that with the increase in the value of the row, the column value also increases
```

```python
sns.heatmap(df.corr()) # to view correlation with color codes
```

## Joint Plot

```python
#Joint Plot is used when we have only two features
```

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='hex') # hex for viewing in hexagon shapes
```

```python
# above shows the relation between total_bill and tip. It also shows the histogram for tip and total_bill
# we can say that majority of the people have given 2$ or 3$ tip and the total_bill was between 10 and 20 most of the times(as it is highly concentrated)
```

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='reg') # reg will draw pdf on histograms and best fit line on the data
```

## Pair Plot

```python
# Pair Plot (also called Scatter Plot) is used when we have more than 2 features
```

```python
sns.pairplot(df) # will draw graphs for all the features (pair wise) that have int or float as the data types
```

```python
sns.pairplot(df, hue='sex') # will show the values of sex for each plot
```

## Dist plot

```python
# Dist Plot is used to check the distribution of the columns feature
```

```python
sns.distplot(df['tip']) # X-axis will contain the tip values and Y-axis will contain the densities
```

```python
sns.distplot(df['tip'], kde=False) # if we want to see the exact counts instead of the densities | kde-kernel density estimation
```
