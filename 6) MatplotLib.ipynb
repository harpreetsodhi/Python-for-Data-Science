```python
import matplotlib.pyplot as plt
# if we write %matplotlib inline then we do not need to write plt.show() for viewing, we can just write plt
%matplotlib inline
```

```python
import numpy as np
```

## Scatter

```python
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
# plt scatter will scatter the values of x and y

plt.scatter(x,y,c='r') # c is the red color
# we do not need to write plt.show() becuse we have already defined %matplotlib inline

plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
plt.savefig('Test.png') # we can aslo right click and save
```

```python
np.pi
```

```python
# plot graph of sine
x = np.arange(0,4*np.pi,0.1) # 0 to 4 and step size 0.1
y = np.sin(x)
plt.plot(x,y)
plt.title('sine wave form')
```

```python
x = np.arange(0,4*np.pi,0.1)

plt.subplot(2,1,1)
plt.plot(x,np.sin(x),'r--')
plt.title('Sine')

plt.subplot(2,1,2)
plt.plot(x,np.cos(x),'g--')
plt.title('Cosine')
```

## Bar Plot

```python
x = [2,8,10]
y = [11,16,9]

x2 = [3,9,11]
y2 = [6,15,7]
```

```python
plt.bar(x,y)
plt.bar(x2,y2,color='g')
plt.title('Bar Graph')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
```

## Histograms

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27]) 
plt.hist(a) # by defaut the bins is 10 which means all the values will be divided into 10 bins(ranges)
plt.title("Histogram")
```

```python
plt.hist(a, bins=20)
plt.title("Histogram")
```

## Box Plot

```python
np.random.randn
np.random.normal
```

```python
data = [np.random.normal(0,std,100) for std in range(1,4)] # Draw random samples from a normal (Gaussian) distribution.
# 0 is the mean and std is 1,2,3 and size is 100, this will return 3 arrays of size 100 each
```

```python
# rectangular box plot -the yellow centre line is 50% percentile, lower and upper edge of the box shows 25% and 75% respectively
plt.boxplot(data, vert=True, patch_artist=True); # vert=True makes the boxes vertical, patch_artist=True is for the color
```

```python
plt.boxplot(data, vert=False);
```

## Pie Chart

```python
labels = ['Python', 'C++', 'Ruby', 'Java']
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1,0,0,0) # this will explode the 1st slice

# plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct= '%1.1f%%', shadow=True) # autopct is used to define the values format
plt.axis('equal') # Set equal scaling (i.e., make circles circular) by changing axis limits
plt.show()
```
